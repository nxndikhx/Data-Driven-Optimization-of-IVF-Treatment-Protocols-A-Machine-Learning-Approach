```python
import pandas as pd
import xlrd
df = pd.read_excel("Clinical parameters Dataset.xlsx")
df
```

```python
# Filter the DataFrame to keep only rows where the "Type of ART" column is 0 (IVF)
df1 = df[df['IVF/ICSI cycle'] == 1]
df1
```

```python
df1 = df.dropna()
df1
```

```python
df1.to_excel('dataset2_for_mp.xlsx', index=False)
df1
```

```python
descriptions = {
    '№': 'Unique identifier for each record',
    'Age': 'Age of the patient',
    'IVF/ICSI cycle': 'The cycle number of IVF/ICSI treatment',
    'FSH, IU/L': 'Follicle Stimulating Hormone level, measured in IU/L',
    'LH, IU/L': 'Luteinizing Hormone level, measured in IU/L',
    'AMH, ng/mL': 'Anti-Müllerian Hormone level, measured in ng/mL',
    'Progesterone, nmol/L': 'Progesterone level, measured in nmol/L',
    'Follicles obtained': 'Number of follicles obtained during the procedure',
    'LH/FSH ratio': 'Ratio of LH to FSH levels',
    'FSHB c.-211 G>T': 'Genetic polymorphism in the FSHB gene',
    'FSHR c.2039 A>G': 'Genetic polymorphism in the FSHR gene'
}

# Create the data dictionary DataFrame
data_dict = pd.DataFrame({
    'Column Name': df1.columns,
    'Data Type': df1.dtypes.values,
    'Description': df1.columns.map(descriptions)
})

# Save the data dictionary to a CSV file
data_dict.to_excel('data_dictionary.xlsx', index=False)

# Print the data dictionary to verify the results
print(data_dict)
```
